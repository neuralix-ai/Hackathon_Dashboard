```python
import os

import pandas as pd
import numpy as np

import plotly.graph_objects as go
```

```python
def plot_stacked_time_series_list(ts_list, width=None, height=None):
    # Initialize a Plotly figure
    fig = go.Figure()

    inset = 0.9
    # Loop over the DataFrame columns, adding each as a separate line in the plot
    for ii in range(len(ts_list)):
        ts = ts_list[ii]
        # Shift the y-values to ensure they don't overlap, according to the given rule
        # y_values = df[column] - i * (df[column].max() - df[column].min() + 1)
        temp_ts = -1*ts
        if temp_ts.max() - temp_ts.min() > 0.00001:
            y_values = ii + inset*(temp_ts - temp_ts.min())/(temp_ts.max() - temp_ts.min())
        else:
            y_values = ii * inset*np.ones([len(temp_ts)])

        

        fig.add_trace(go.Scatter(y=y_values, mode='lines', name=f"time series {ii}"))


    # Update layout to improve readability
    fig.update_layout(
        title='Stacked Time Series Plot',
        xaxis_title='Time',
        yaxis_title='Value',
        yaxis_autorange='reversed',  # This reverses the y-axis
        legend_title='Time Series',
        hovermode='x unified',
        width=width,  # Width in pixels
        height=height,  # Height in pixels
    )

    # Show the plot
    fig.show()

    return None
```

```python
ts_length = 10000
ts_list = [np.random.uniform(0,1,ts_length) for i in range(4)]
plot_stacked_time_series_list(ts_list, width=None, height=None)
```
